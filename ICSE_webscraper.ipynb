```python
# ✅ 1. Upload and unzip your ZIP of PDFs
from google.colab import files
import zipfile
import os

uploaded = files.upload()
zip_name = list(uploaded.keys())[0]

extract_path = "pdfs"
os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Extracted folders/files:", os.listdir(extract_path))
```

```python
# ✅ 2. Recursively find all PDF files (even in subfolders)
from pathlib import Path

pdf_paths = list(Path(extract_path).rglob("*.pdf"))
print(f"✅ Found {len(pdf_paths)} PDF files.")

for path in pdf_paths:
    print("📄", path.name)
```

```python
# ✅ 3. Define reproducibility score rubric
import fitz  # PyMuPDF
import pandas as pd

def score_from_cues(text, cues):
    text = text.lower()
    return min(5, sum(cue in text for cue in cues))

def score_paper(text):
    return {
        "Code Availability": score_from_cues(text, ["github.com", "code available", "open-source", "repository"]),
        "Dataset Availability": score_from_cues(text, ["dataset", "data available", "benchmark", "codesearchnet"]),
        "Documentation Quality": score_from_cues(text, ["readme", "setup", "installation", "documentation"]),
        "GPU Requirements": score_from_cues(text, ["gpu", "tesla", "nvidia", "cuda"]),
        "Computer Requirements": score_from_cues(text, ["cpu", "ram", "memory", "system specs"]),
        "Result Reproducibility": score_from_cues(text, ["reproducibility", "reproduce", "replication", "replicable"])
    }
```

```python
# ✅ 4. Score each PDF
results = []

for pdf_path in pdf_paths:
    try:
        with fitz.open(pdf_path) as doc:
            text = "".join([page.get_text() for page in doc])
        scores = score_paper(text)
        scores["Filename"] = pdf_path.name
        results.append(scores)
    except Exception as e:
        print(f"⚠️ Skipping {pdf_path.name}: {e}")

df = pd.DataFrame(results)
df = df[["Filename", "Code Availability", "Dataset Availability", "Documentation Quality",
         "GPU Requirements", "Computer Requirements", "Result Reproducibility"]]

print("✅ Done scoring. Here's your scorecard:")
df.head(10)
```

```python
# ✅ Calculate average score per paper
df["Average Score"] = df[
    ["Code Availability", "Dataset Availability", "Documentation Quality",
     "GPU Requirements", "Computer Requirements", "Result Reproducibility"]
].mean(axis=1)

# ✅ Sort and show top 3 most reproducible papers
top_3 = df.sort_values("Average Score", ascending=False).head(3)

print("🏆 Top 3 Most Reproducible Papers:")
top_3_display = top_3[["Filename", "Average Score"]]
display(top_3_display)
```

```python
import matplotlib.pyplot as plt

# Sort by average score for clearer graph
df_sorted = df.sort_values("Average Score", ascending=False)

# Plot
plt.figure(figsize=(12, 6))
plt.barh(df_sorted["Filename"], df_sorted["Average Score"], color="skyblue")
plt.xlabel("Average Reproducibility Score (0–5)")
plt.title("📊 Reproducibility Score per Paper")
plt.gca().invert_yaxis()  # highest score at top
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
